```python
import time
import requests
import os
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def write_json(path, data):
    with open(path, 'w', encoding="utf-8") as f:
        json.dump(data, f, indent=2)
```

```python
def download(filename, url):
    # Do not download again if the file already exists
    # Use the appropriate os module function to check if the target file already exists
    if os.path.exists(filename):
        return (str(filename) + " already exists!")
    # make the request
    r = requests.get(url)
    r.raise_for_status()
    
    # get the text from the request
    text = r.text
    
    # open the file
    f = open(filename, 'w', encoding="utf-8")
        
        
    # write the text to the file
    f.write(text)
    
    # close the file
    f.close
    return (str(filename) + " created!")
```

```python
url_19 = 'https://raw.githubusercontent.com/msyamkumar/cs220-s22-projects/main/p12/2019-2020.html'
url_20 = 'https://raw.githubusercontent.com/msyamkumar/cs220-s22-projects/main/p12/2020-2021.html'
url_21 = 'https://raw.githubusercontent.com/msyamkumar/cs220-s22-projects/main/p12/2021-2022.html'
download('2019-2020.html', url_19)
download('2020-2021.html', url_20)
download('2021-2022.html', url_21)
```

```python
def parse_html(html_file):
    f = open(html_file)
    year_str = f.read()
    f.close()

    bs_obj = BeautifulSoup(year_str, "html.parser")
    
    table = bs_obj.find("table") # works only when you have exactly 1 table
    header = [th.get_text() for th in table.find_all('th')]

    year_data = []

    tr_elements = table.find_all('tr')
    for tr in tr_elements[1:]: # Skip first row (header row)
        row = {}
        row["Year"] = html_file[0:9]
        td_elements = tr.find_all('td')
        for idx in range(len(td_elements)):
            td = td_elements[idx]
            val = td.get_text()      
            if val == "-":
                row[header[idx]] = None
            elif header[idx] in ["World Rank", "National Rank"]:
                row[header[idx]] = int(val)
            elif header[idx] in ["Institution", "Country"]:
                row[header[idx]] = str(val)
            elif header[idx] in ["Quality of Education Rank", "Alumni Employment Rank", "Quality of Faculty Rank", "Research Performance Rank", "Score"]:
                row[header[idx]] = float(val)
            else:
                row[header[idx]] = val
        year_data.append(row)   
    return year_data
```

```python
assert(parse_html("2019-2020.html")[-1]["Institution"] == 'Government College University Faisalabad')
assert(parse_html("2020-2021.html")[15]["Score"] == 89.0)
assert(parse_html("2021-2022.html")[100]["Country"] == 'United Kingdom')
assert(parse_html("2021-2022.html")[25]["World Rank"] == 26)
assert(parse_html("2020-2021.html")[-5]["National Rank"] == 15)
assert(parse_html("2019-2020.html")[50]["Quality of Faculty Rank"] == 78)
assert(parse_html("2021-2022.html")[87]["Alumni Employment Rank"] == 464)
assert(parse_html("2020-2021.html")[40]["Research Performance Rank"] == 398)
assert(parse_html("2019-2020.html")[0]["Year"] == "2019-2020")
```
