```python
import datetime
import csv
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data

csv_rows = process_csv("vaccinations.csv")
csv_header = csv_rows[0]
csv_data = csv_rows[1:]

def cell(row_idx, col_name):
    col_idx = csv_header.index(col_name)
    val = csv_data[row_idx][col_idx]
    if val == "":
        return None
    list_int = ["daily_vaccinations", "total_vaccinations", "people_vaccinated", "people_fully_vaccinated", "total_boosters", "population"]
    if col_name in list_int:
        val = int(val)
    list_string = ["date", "country"]
    if col_name in list_string:
        return str(val)
    return val
```

```python
#Q1: What dates are mentioned in the dataset?
dates = []
for row_idx in range(len(csv_data)):
    if cell(row_idx, "date") not in dates:
        dates.append(cell(row_idx, "date"))
all_dates = dates
all_dates
```

```python
#Q2: What countries are listed in the dataset?
countries = []
for row_idx in range(len(csv_data)):
    if cell(row_idx, "country") not in countries:
        countries.append(cell(row_idx, "country"))
countries
```

```python
#Q3: What is the population of each country in the dataset?
country_pop = {}
for row_idx in range(len(csv_data)):
    country_name = cell(row_idx, "country")
    population = cell(row_idx, "population")
    if country_name not in country_pop:
        country_pop[country_name] = None
    if country_name in country_pop:
        if population == None:
            continue
        country_pop[country_name] = population
country_pop
```

```python
#Q4: How many countries have an entry for daily vaccination on Jan 28, 2022?
count = 0
for row_idx in range(len(csv_data)):
    if cell(row_idx, "date") == "01/28/2022":
        if cell(row_idx, "daily_vaccinations") != None:
            count += 1
count
```

```python
def daily_vaccinations_on(search_date):
    num_vaxx = {}
    for row_idx in range(len(csv_data)):
        data_date = cell(row_idx, "date")
        data_country_name = cell(row_idx, "country")
        data_daily_vaxx = cell(row_idx, "daily_vaccinations")
        if data_date == search_date:
            if data_country_name not in num_vaxx:
                num_vaxx[data_country_name] = None
            if data_country_name in num_vaxx:
                num_vaxx[data_country_name] = data_daily_vaxx
    return num_vaxx
```

```python
#Q5: For each country, how many vaccinations were given on Jan 31, 2022?
daily_vaccinations_on("01/31/2022")
```

```python
#Q6: For each country, what is the sum of the vaccinations given on all of the dates mentioned in the dataset?
sum_vacc_dict = {}
for i in all_dates:
    full_data = daily_vaccinations_on(i)
    for country in full_data:
        if country not in sum_vacc_dict:
            sum_vacc_dict[country] = full_data[country]
        elif full_data[country] != None:
            if sum_vacc_dict[country] == None:
                sum_vacc_dict[country] = full_data[country]
            else:
                sum_vacc_dict[country] += full_data[country]
sum_vacc_dict
```

```python
#Q7: Which country completed the most number of vaccinations during the dates mentioned in the dataset?
max_value = 0
key_list = list(sum_vacc_dict.keys())
value_list = list(sum_vacc_dict.values())
for value in sum_vacc_dict.values():
    if value == None:
        continue
    if value > max_value:
        max_value = value
        max_key = key_list[value_list.index(max_value)]
max_key
```

```python
#Q8: How many vaccinations (across the world) were given on each date?
total_vaxx = {}
for row_idx in range(len(csv_data)):
    date = cell(row_idx, "date")
    vacc = cell(row_idx, "daily_vaccinations")
    if vacc != None:
        if date not in total_vaxx:
            total_vaxx[date] = vacc
        else:
            if vacc != None:
                total_vaxx[date] += vacc
total_vaxx
```

```python
#Q9: For each country in the dataset, when is the last date for which we have information about the number of fully vaccinated people?
fully_vaxx = {}
for row_idx in range(len(csv_data)):
    country = cell(row_idx, "country")
    vaxx = cell(row_idx, "people_fully_vaccinated")
    date = cell(row_idx, "date")
    if country not in fully_vaxx:
        fully_vaxx[country] = None
    if country in fully_vaxx:
        if vaxx == None:
            continue
        if vaxx != None:
            fully_vaxx[country] = date
fully_vaxx        
```

```python
def get_number_of_days(start_date, end_date):
    day1 = datetime.datetime.strptime(start_date, '%m/%d/%Y') 
    day2 = datetime.datetime.strptime(end_date, '%m/%d/%Y')
    delta = day2 - day1
    return delta.days

def most_recent_total(col_name, given_date):
    country_info = {}
    for row_idx in range(len(csv_data)):
        country = cell(row_idx, "country")
        date = cell(row_idx, "date")
        col_value = cell(row_idx, col_name)
        diff_day = get_number_of_days(date, given_date)
        if country not in country_info:
            country_info[country] = None
        if col_value != None and diff_day >= 0:  # Make sure the column value is not missing AND Check if the date is on or before given_date:
            country_info[country] = col_value
    return country_info
```

```python
#Q10: How many people were estimated to be fully vaccinated in each country, as of 31 Jan 2022?
people_fully_vaxx = most_recent_total("people_fully_vaccinated", "01/31/2022")
people_fully_vaxx
```

```python
#Q11: Among countries with population less than 100 million, which country is estimated to have had the highest number of fully vaccinated people by 31 Jan 2022?
max_value = 0 
value_vaxx = list(people_fully_vaxx.values())
key_vaxx = list(people_fully_vaxx.keys())  
for country in people_fully_vaxx:
    for country in country_pop:
        if country_pop[country] == None:
            continue
        if country_pop[country] < 100000000:
            value = people_fully_vaxx[country]
            if value == None:
                continue
            if value > max_value:
                max_value = value 
                max_key = key_vaxx[value_vaxx.index(max_value)]
max_key
```

```python
vaccination_stats = {}

for date in all_dates:
    daily_vax_dict = daily_vaccinations_on(date)
    total_vax_dict = most_recent_total("total_vaccinations", date)
    people_vax_dict = most_recent_total("people_vaccinated", date) 
    fully_vax_dict = most_recent_total("people_fully_vaccinated", date) 
    total_booster_dict = most_recent_total("total_boosters", date)
    
    for country in countries:
        if country not in vaccination_stats:
            vaccination_stats[country] = {}
        vaccination_stats[country][date] = {}
        vaccination_stats[country][date]['country'] = country
        vaccination_stats[country][date]['date'] = date
        vaccination_stats[country][date]['daily_vaccinations'] = daily_vax_dict[country]
        vaccination_stats[country][date]['total_vaccinations'] = total_vax_dict[country]
        vaccination_stats[country][date]['people_vaccinated'] = people_vax_dict[country]
        vaccination_stats[country][date]['people_fully_vaccinated'] = fully_vax_dict[country]
        vaccination_stats[country][date]['total_boosters'] = total_booster_dict[country]
        vaccination_stats[country][date]['population'] = country_pop[country]
        
```

```python
#Q15: What are the vaccination records for the country with the highest population in the dataset on 29 Jan 2022?
max_pop = 0
max_country = None
for country in country_pop:
    if country_pop[country] == None:
        continue
    if country_pop[country] > max_pop:
        max_pop = country_pop[country]
        max_country = country
vaccination_stats[max_country]["01/29/2022"]
```

```python
#Q16: What is the estimated ratio of people who have been partly vaccinated to people who have been fully vaccinated in each country, as of Jan 31, 2022?
vaxx_ratio = {}
# party_vaxx_ppl = 0
# fully_vaxx_ppl = 
# people partly vaccinated = people fully vaccinated - people vaccinated 
                                         

for country in countries:
    if vaccination_stats[country]["01/31/2022"]['people_fully_vaccinated'] == None or vaccination_stats[country]["01/31/2022"]['people_vaccinated'] == None:
        vaxx_ratio[country]= None
        continue
    ratio = (vaccination_stats[country]["01/31/2022"]['people_vaccinated'] - vaccination_stats[country]["01/31/2022"]['people_fully_vaccinated']) / vaccination_stats[country]["01/31/2022"]['people_fully_vaccinated']
    vaxx_ratio[country] = ratio
vaxx_ratio 
```

```python
#Q17: Which country has administered the most number of booster shots as of Jan 31, 2022
most_boost = 0
most_boost_country = None
for country in countries:
    if vaccination_stats[country]["01/31/2022"]['total_boosters'] == None:
        continue
    if vaccination_stats[country]["01/31/2022"]['total_boosters'] > most_boost:
        most_boost = vaccination_stats[country]["01/31/2022"]['total_boosters']
        most_boost_country = country
most_boost_country
```

```python
#Q18: Which country is estimated to have administered the most number of booster shots per capita, as of 25 Jan 2022?
most_boost_per_capita = 0
most_boost_per_country = None
for country in countries:
    if vaccination_stats[country]["01/25/2022"]['total_boosters'] == None or vaccination_stats[country]["01/25/2022"]['population'] == None:
        continue
    ratio = vaccination_stats[country]["01/25/2022"]['total_boosters'] / vaccination_stats[country]["01/25/2022"]['population']
    if ratio > most_boost_per_capita:
        most_boost_per_capita = ratio
        most_boost_per_country = country
most_boost_per_country
```

```python
#Q19: What is the percentage of population boosted by Jan 31, 2022 of each country in the dataset?
pop_boosted = {}

for country in countries:
    if vaccination_stats[country]["01/31/2022"]['total_boosters'] == None or vaccination_stats[country]["01/31/2022"]['population'] == None:
        continue
    boosted_percent = (vaccination_stats[country]["01/31/2022"]['total_boosters'] / vaccination_stats[country]["01/31/2022"]['population'])*100
    pop_boosted[country] = boosted_percent
pop_boosted     
```
