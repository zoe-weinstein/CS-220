```python
def bar_plot(df, x, y):
    ax = df.plot.bar(color = 'black', x=x, y=y, legend = False)
    ax.set_xlabel(x)
    ax.set_ylabel(y)

def scatter_plot(df, x, y):
    scatter = df.plot.scatter(x=x, y=y, color = 'black')

def plot_horizontal_bar(df, x):
    df = df.set_index(x)
    ax = df.plot.barh()
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.9))

def plot_pie(df, x, y, title=None):
    df = df.set_index(x)
    ax = df.plot.pie(y=y, legend=False)
    ax.set_ylabel(None)
    ax.set_title(title)

def get_regression_coeff(df, x, y):
    df["1"] = 1
    res = np.linalg.lstsq(df[[x, "1"]], df[y], rcond=None)
    coefficients = res[0]
    m = coefficients[0]
    b = coefficients[1]
    return (m, b)

def plot_regression_line(df, x, y):
    m = get_regression_coeff(df, x, y)[0]
    b = get_regression_coeff(df, x, y)[1]
    ax = df.plot.scatter(x = x, y = y, color = 'black')
    df['fit'] = (m*df[x])+b
    df.plot.line(ax=ax, x=x, y= 'fit', color = 'red')
```

```python
def write_json(path, data):
    with open(path, 'w', encoding="utf-8") as f:
        f.write(data)
```

```python
def download(filename, url):
    if os.path.exists(filename):
        return (str(filename) + " already exists!")

    r = requests.get(url)
    r.raise_for_status()

    write_json(filename, r.text)
    
    return (str(filename) + " created!")

download("QSranking.json","https://raw.githubusercontent.com/msyamkumar/cs220-s22-projects/main/p13/QSranking.json")
```

```python
df1 = pd.read_json('QSranking.json')
conn = sqlite3.connect('rankings.db')
df1.to_sql("rankings", conn, if_exists="replace", index=False)
pd.read_sql("select * from 'rankings'", conn)
```

```python
#Q1: Which 10 institutions in Canada had the highest number of international students in 2020?
query = """
select institution_name, international_students
from rankings
where country = 'Canada' and year = 2020
order by international_students desc
limit 10
"""

pd.read_sql(query, conn)
```

```python
#Q2: Which 10 institutions in the United States had the highest reputation in 2019?

query = """
select institution_name, (academic_reputation + employer_reputation) AS reputation
from rankings
where country = 'United States' and year = 2019
order by reputation desc, institution_name 
limit 10 
"""

pd.read_sql(query, conn)
```

```python
#Q3: Which 10 countries had the most institutions listed in the year 2020?
query = """
select country, count(institution_name) AS num_of_institutions
from rankings
where year = 2020
group by country
order by num_of_institutions desc, country
limit 10 
"""

pd.read_sql(query, conn)
```

```python
#Q4: Plot the data from Q3 as a bar chart, along with the sum of all other countries' institutions.
query = """
select country, count(institution_name) AS num_of_institutions
from rankings
where year = 2020
group by country
order by num_of_institutions desc, country
"""

df = pd.read_sql(query, conn)
top10 = df[:10]
country_institutions = top10.append({"country": "Other", "num_of_institutions": df['num_of_institutions'].iloc[10:].sum() }, ignore_index=True)
bar_plot(country_institutions, x = "country", y= "num_of_institutions")

```

```python
#Q11: Find the correlation between international_students and overall_score for institutions in the'United Kingdom' that were ranked in the top 100 in the year 2020. Perform the same calculation forinstitutions in the 'United States'. What is the difference between these two correlations?

query = """
select international_students, overall_score
from rankings
where country = 'United Kingdom' and year = 2020 and rank <= 100

"""
uk = pd.read_sql(query, conn)
uk_correlation = uk.corr().iloc[0, 1]

query = """
select international_students, overall_score
from rankings
where country = 'United States' and year = 2020 and rank <= 100

"""
us = pd.read_sql(query, conn)
us_correlation = us.corr().iloc[0, 1]

difference = uk_correlation - us_correlation
difference
```
